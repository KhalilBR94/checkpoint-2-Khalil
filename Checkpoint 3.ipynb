```python
import pandas as pd
data=pd.read_csv(r'F:\SAUVGARDE 2020\user\Desktop\Gomycode\checkpoint 3\titanic-passengers.csv',sep=';')
data

```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
data
```

```python
data.Age = data.Age.astype(int)
```

```python
data.drop(['PassengerId', 'Ticket','Cabin'], axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Survived']=encoder.fit_transform(data['Survived'])
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Sex']=encoder.fit_transform(data['Sex'])
data
```

```python
data
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
data
```

```python
one_hot=pd.get_dummies(data['Embarked'])
data=data.drop('Embarked',axis=1)
data=data.join(one_hot)
```
