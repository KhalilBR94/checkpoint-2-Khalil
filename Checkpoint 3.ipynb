```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map ( data ):
    corr = data.corr()   
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }   

        )   

plot_correlation_map(data)
       #la heatmap nous montre la corrélation entre les variables de l'ensemble de nos données. 
    #La corrélation entre les variables va de -1 à 1. Si la corrélation est négative entre deux variables 
    #ca veut dire que si l'une des deux augmente l'autre diminue.
    #Une valeur de zéro indique qu'il n'y a aucune corrélation entre les variables.
    
    
```
